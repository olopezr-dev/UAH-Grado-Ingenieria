<h1>Empaquetado de archivos</h1>

<div style="font-size: 17px">
Dos de las herramientas más populares en Linux para la compresión y empaquetado de archivos son <code>tar</code> y <code>gzip</code>. Aunque cumplen funciones diferentes, a menudo se usan en conjunto para agrupar y comprimir archivos eficientemente:

  - **<code>tar</code>**: Empaqueta varios archivos y directorios en un único archivo sin comprimir, manteniendo su estructura de directorios. Con la opción <code>-f</code> indicamos el nombre del archivo. Se puede ejecutar con varios modos:
    - **<code>--create</code> (<code>-c</code>)**: Crea un <code>.tar</code> nuevo.
    - **<code>--extract</code> (<code>-x</code>)**: Extrae los archivos de un <code>.tar</code> existente.
    - **<code>--list</code> (<code>-t</code>)**: Muestra una lista de los archivos del <code>.tar</code> especificado. Si se añade la opción <code>-v</code> al crear o extraer archivos se obtiene la misma salida.

**Ejemplo**: Se crea una estructura de archivos, se empaqueta en el archivo <code>prueba.tar</code> y se comprueba su contenido con la opción <code>-t</code>:
</div>

```python
%%shell
#Creación de la estructura de directorios y archivos a empaquetar
mkdir directorio
mkdir directorio/subdirectorio
echo "Primer archivo" > directorio/prueba1.txt
echo "Segundo archivo" > directorio/subdirectorio/prueba2.txt
#Creación del archivo .tar
tar -cf prueba.tar directorio/
tar -tf prueba.tar
#Borrar los archivos resultantes
rm directorio/prueba1.txt
rm directorio/subdirectorio/prueba2.txt
rm prueba.tar
rmdir directorio/subdirectorio
rmdir directorio
```

<div style="font-size: 17px">

  - **<code>gzip</code>**: Comprime archivos individuales, reduciendo el espacio que ocupan en el sistema. Puede comprimir archivos <code>.tar</code>.
  - **<code>gunzip</code>**: Descomprime archivos de extensión <code>.gz</code>.

**Ejemplo**: Se crea un archivo <code>prueba.tar</code> a partir de un directorio, se comprime con <code>gzip</code>, y descomprime de vuelta y se comprueba su contenido:
</div>

```python
%%shell
#Creación de la estructura de directorios y archivos a empaquetar
mkdir directorio
echo "Prueba" > directorio/prueba.txt
#Creación del archivo .tar
tar -cf prueba.tar directorio/
echo "Archivo .tar creado:"
ls -l | grep prueba
#Compresión en formato .gz
gzip prueba.tar
echo "Comprimido en .gz:"
ls -l | grep prueba
#Borrar los archivos resultantes
rm directorio/prueba.txt
rmdir directorio
#Descomprimir
gunzip prueba.tar.gz
echo "Descomprimido:"
ls -l | grep prueba
#Borrar los archivos resultantes
rm prueba.tar
```

<div style="font-size: 17px">

**Ejemplo**: Existe una opción en la orden <code>tar</code> (<code>-z</code>) que comprime los archivos directamente (genera el archivo <code>.tar</code> y lo comprime, sin borrar el <code>.tar</code>):
</div>

```python
%%shell
#Creación de la estructura de directorios y archivos a empaquetar
mkdir directorio
echo "Prueba" > directorio/prueba.txt
#Empaquetado y compresión directa
tar -czf prueba.tar.gz directorio/
echo "Archivo empaquetado y comprimido:"
ls -l | grep prueba
#Borrar los archivos resultantes
rm directorio/prueba.txt
rmdir directorio
#Descomprimir
gunzip prueba.tar.gz
echo "Descomprimido:"
ls -l | grep prueba
#Borrar los archivos resultantes
rm prueba.tar
```
